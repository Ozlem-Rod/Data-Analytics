```python
# Gender information
sex = df.groupby('gender').size()
print(sex)
```

```python
# Gender by Age Analysis
age = df.groupby(['gender','age']).size()
print(age)
```

```python
# Average Number of Medication by Gender and Age Analysis
by_sex = df.groupby(['gender','age'])['num_medications'].mean()
print(by_sex)
```

```python
# Diabetes medication Analysis by Gender
by_medicine = df.groupby(['gender','diabetesMed']).size()
print(by_medicine)
```

```python
# Hospital Readmission Analysis by Gender
by_readmit = df.groupby(['readmitted','gender']).size()
print(by_readmit)
```

```python
# Hospital Readmission Analysis by those taking medication
by_med = df.groupby(['gender','diabetesMed','readmitted']).size()
print(by_med)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cleaning task to map counts for graph
by_med = df.groupby(['gender','diabetesMed','readmitted']).size().reset_index(name='count')
print(by_med)
```

```python
# Bar chart for Diabetes Medication Impact on Hospital Readmission
plt.figure(figsize=(10, 6))
sns.barplot(data=by_med, x='readmitted', y='count', hue='diabetesMed', ci=None)
plt.xlabel('Readmission Status')
plt.ylabel('Count')
plt.title('Diabetes Medication Impact on Readmission')
plt.legend(title='Diabetes Medication')
plt.show()
```

```python
# Analysis for Changes in Glucose due to Medication
by_change = df.groupby(['gender','diabetesMed','change']).size()
print(by_change)
```

```python
# Analysis by Gender for Diabetes Medication, Changes to Glucose and Readmission Rates
by_changed = df.groupby(['gender','diabetesMed','change','readmitted']).size()
print(by_changed)
```
